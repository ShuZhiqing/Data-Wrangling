### 3.1 Modify 'ContractType' and 'ContractTime'

Give a look to the values in `ContractType` and `ContractTime` in both two datasets.

```python
dataset1.ContractType.value_counts()
```

```python
dataset2['Contract Type'].value_counts()
```

```python
dataset2['Contract Time'].value_counts()
```

Basically, values in dataset2 is the abbreviation of their counterpart in dataset2. Since we are required to adapt the schema in dataset1 for the global schema as much as we could, so we need to replace the abbreviation with its full name. Missing Contract Time and Contract Type in dataset2 can be understood as `non-specified`, so just fill in 'non-specified'.

```python
# permanent, contract or non-specified.
dataset2['Contract Time'].replace({'perm.':'permanent','contr.':'contract','':'not available'},inplace = True)
dataset2['Contract Time'].fillna('non-specified',inplace=True)
```

```python
dataset2['Contract Type'].replace({'pt':'part_time','ft':'full_time','':'not available'},inplace = True)
dataset2['Contract Type'].fillna('non-specified',inplace=True)
```

Check the result.

```python
dataset2['Contract Time'].value_counts()
```

```python
dataset2['Contract Type'].value_counts()
```

### 3.3 Modify 'Salary'

We need to modify `Salary per month` in dataset2 into `Salary per annum` by following code.

```python
dataset2['Salary per month'] = dataset2['Salary per month'].apply(lambda x: x * 12)
```

Renaming `Salary per month` as `Salary per annum`.

```python
dataset2 = dataset2.rename(columns ={'Salary per month':'Salary per annum'})
```

Modify the data type of `Salary per annum` from `int` to `str`.

```python
dataset2[['Salary per annum']] = dataset2[['Salary per annum']].astype(str)
```

### 3.4 Modify 'OpenDate' and 'CloseDate'

```python
dataset1.head(1)
```

```python
dataset2.head(1)
```

We can find in dataset1, `OpenDate`	and `CloseDate` is in format as `20121103T000000`, while in dataset2, they are recorded as datatime.

Modify the format of `OpenDate` and `CloseDate` in dataset2 to `"%Y%m%dT%H%M%S"`.

```python
dataset2['OpenDate']= pd.to_datetime(dataset2.OpenDate,format="%Y-%m-%d %H:%M:%S").\
                        apply(lambda x: x.strftime("%Y%m%dT%H%M%S"))
```

```python
dataset2['CloseDate']= pd.to_datetime(dataset2.CloseDate,format="%Y-%m-%d %H:%M:%S").\
                        apply(lambda x: x.strftime("%Y%m%dT%H%M%S"))
```

Check the result.

```python
dataset2.head(1)
```

#### 4.2 Find a proper global key for the integrated job data

I choose `Title`,`Location`,`Salary per annum`,`Company`,`OpenDate` as the global key for the integrated job data based on following reason:
* Some job title may be named very generally, such as Product Manager, we are not able to make sure if it is duplicted record just by name. (Although in our dataset, title value is unique in both two dataset, in real life, it seems to be impossible.)
* Company is likely to be unique, so add it with title can help us to find duplications.
* Same title or position may exist in one company, and one compamny may have branch in different place, so location is also very necessary.
* While same position in one branch may has different salary, and salary may indicate the ContractType since full time is normally paid higher than partime.
* One company is likely to recruit same position at same salary in same branch in different time, so including Opendate will be necessary.
